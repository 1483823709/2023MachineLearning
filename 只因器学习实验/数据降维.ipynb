```python
# 导入需要的库
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn.decomposition import PCA

# 导入数据集
iris = datasets.load_iris()
X = iris.data
y = iris.target

# 创建PCA模型，指定要降到的维度为两维度，，为了方便画图
JiangWei = PCA(n_components=2)

# 用数据矩阵来训练模型
JiangWei.fit(X)

# 对数据矩阵进行降维变换，得到降维后的数据矩阵
X_new = JiangWei.transform(X)

# 查看降维后的数据矩阵的形状
print("Shape", X_new.shape)

# 查看模型的方差解释率，即每个主成分保留了多少原始数据的信息量，应该是一个数组，元素和为1
print("Var", JiangWei.explained_variance_ratio_)

# 绘制降维后的数据点的散点图，用不同颜色表示不同类别
plt.scatter(X_new[:, 0], X_new[:, 1], c=y)
plt.xlabel('First Principal Component')
plt.title('Result')
plt.show()
```
